```python
def get_types(df, samples_amount=3):
    num_columns = list(df.select_dtypes(include='number').columns)
    cat_columns = [column for column in df.columns if column not in num_columns]

    print(f'{cat_columns=}', f'{num_columns=}', sep='\n\n')

    rows = df.sample(samples_amount)
    column_types = {column: ['number'] for column in num_columns}
    column_types.update({column: ['category'] for column in cat_columns})
    column_types = pd.DataFrame(column_types)

    column_dtypes = dict(zip(df.columns, [df[column].dtype for column in df.columns]))
    column_dtypes = {column: [dtype] for column, dtype in column_dtypes.items()}
    column_dtypes = pd.DataFrame(column_dtypes)
    
    display(pd.concat([column_dtypes, column_types, rows]))
    return num_columns, cat_columns

num_columns, cat_columns = get_types(X_train, 3)
```

```python
def get_str_values_from_number_series(series):
    str_values = list()
    for value in series:
        try:
            float(value)
        except:
            str_values.append(value)
    return list(set(str_values))
    
train_str_values = get_str_values_from_number_series(X_train['tempo'])
test_str_values = get_str_values_from_number_series(X_test['tempo'])
```
